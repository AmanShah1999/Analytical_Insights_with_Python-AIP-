# gaussian_filter

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# Assuming 'data' is your DataFrame
data = pd.read_csv("RED MUD SA_XRD.txt", sep='\t')
#print(data.head())

# Apply Gaussian smoothing to the 'X' column
smoothed_x = gaussian_filter1d(data['Acquisition'], sigma=5)

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Plotting original data
axes[0].plot(data['X'], data['Acquisition'], linestyle='-', color="r")
axes[0].set_ylabel('Acquisition')
axes[0].set_title('Original Data')
#axes[0].grid(True)
axes[0].set_xlim(20, 80)

# Plotting smoothed data
axes[1].plot(data['X'], smoothed_x, linestyle='-', color="b")
axes[1].set_xlabel('X')
axes[1].set_ylabel('Acquisition')
axes[1].set_title('Smoothed Data')
#axes[1].grid(True)
axes[1].set_xlim(20,80)

plt.tight_layout()
plt.show()
```

# Savitzky-Golay smoothing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Assuming 'data' is your DataFrame
data = pd.read_csv("RED MUD SA_XRD.txt", sep='\t')

# Apply Savitzky-Golay smoothing to the 'Acquisition' column
smoothed_acquisition = savgol_filter(data['Acquisition'], window_length=20, polyorder=3)

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Plotting original data
axes[0].plot(data['X'], data['Acquisition'], linestyle='-', color="r")
axes[0].set_ylabel('Acquisition')
axes[0].set_title('Original Data')
axes[0].set_xlim(20, 80)

# Plotting smoothed data using Savitzky-Golay filter
axes[1].plot(data['X'], smoothed_acquisition, linestyle='-', color="b")
axes[1].set_xlabel('X')
axes[1].set_ylabel('Acquisition')
axes[1].set_title('Smoothed Data (Savitzky-Golay)')
axes[1].set_xlim(20, 80)

plt.tight_layout()
plt.show()
```
